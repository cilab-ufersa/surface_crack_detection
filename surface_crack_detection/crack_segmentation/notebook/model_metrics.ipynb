### **1. Importing necessary package**

```python
import pandas as pd
```

### **2. Loading the metrics of models**

```python
paths = ['../output/crack_detection_4.out', "../output/crack_detection_2.out"]

unet_mobilenet = pd.read_csv(paths[0], delimiter=';')
fpn_inceptionv3 = pd.read_csv(paths[1], delimiter=';')
```

### **3. Showing up the results**

#### 3.1. Unet-mobilenet

```python
unet_mobilenet
```

```python
precision = unet_mobilenet['Precision'].mean()
recall = unet_mobilenet['Recall'].mean()
f1_score = unet_mobilenet['F1_score'].mean()
accuracy = unet_mobilenet['accuracy'].mean()
val_accuracy = unet_mobilenet['val_accuracy'].mean()
loss = unet_mobilenet['loss'].mean()
val_loss = unet_mobilenet['val_loss'].mean()

print("====== unet's metrics ======")
print(f"Precision: {(precision * 100):.2f}%")
print(f"Recall: {(recall * 100):.2f}%")
print(f"F1_score: {(f1_score * 100):.2f}%")
print(f"accuracy: {(accuracy * 100):.2f}%")
print(f"val_accuracy: {(val_accuracy * 100):.2f}%")
print(f"loss: {(loss):.2f}")
print(f"val_loss: {(val_loss):.2f}")
```
